# 인구피라미드
1. 데이터 읽어오기

```python
import pandas as pd

df_m = pd.read_excel('data/2509인구현황.xlsx', skiprows=3,
                    index_col='행정기관', usecols='B,E:Y')

df_m.head()
```

```python
# 데이터를 문자열에서 정수로 변환해주기

print(df_m.dtypes)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
```

```python
df_m.head()
```

```python
df_m.head()
```

```python
df_w = pd.read_excel('data/2509인구현황.xlsx', skiprows=3,
                    index_col='행정기관', usecols='B,AB:AV')

df_w.head()
```
