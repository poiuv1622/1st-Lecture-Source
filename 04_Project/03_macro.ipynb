```python
import pyautogui
import time

# 각 요소의 좌표를 직접 측정하여 입력하세요!
start_year_pos = (965, 74343)   # 시작년도 클릭 위치
start_month_pos = (1090, 7434)  # 시작월 클릭 위치
end_year_pos = (1260, 7413)     # 종료년도 클릭 위치
end_month_pos = (1408, 7413)    # 종료월 클릭 위치
search_btn_pos = (1480, 1006)   # 검색버튼 클릭 위치
download_btn_pos = (1590, 1000) # 엑셀다운로드 버튼 위치

time.sleep(2) # 시작 전 준비 시간

# 1. 시작 년도 클릭 및 입력
pyautogui.click(*start_year_pos)
time.sleep(0.5)
pyautogui.write('2023')
pyautogui.press('enter')
time.sleep(0.5)

# 2. 시작 월 클릭 및 입력
pyautogui.click(*start_month_pos)
time.sleep(0.5)
pyautogui.write('01')
pyautogui.press('enter')
time.sleep(0.5)

# 3. 종료 년도 클릭 및 입력
pyautogui.click(*end_year_pos)
time.sleep(0.5)
pyautogui.write('2025')
pyautogui.press('enter')
time.sleep(0.5)

# 4. 종료 월 클릭 및 입력
pyautogui.click(*end_month_pos)
time.sleep(0.5)
pyautogui.write('06')
pyautogui.press('enter')
time.sleep(0.5)

# 5. 검색 버튼 클릭 후 대기
pyautogui.click(*search_btn_pos)
time.sleep(4) # 데이터 로딩 대기

# 6. 엑셀 다운로드 클릭 + 엔터 + 대기
pyautogui.click(*download_btn_pos)
time.sleep(0.5)
pyautogui.press('enter')
time.sleep(4) # 다운로드 대기

print("매크로 완료")
```
